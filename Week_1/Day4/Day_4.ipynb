```python
def tax_cal():
    actual = float(input('Enter the actual amount of your property: '))
    assess = actual * 0.6
    tax = round(assess * 0.0072, 2)
    print(f'Assessment Value: {assess}\nTax Amount: {tax}')  

tax_cal()                 
```

```python
def paint_cal():
    wall = float(input('Enter the size of wall in square feet: '))
    paint_price = float(input('Enter the price of paint per gallon: '))
    
    gallon = round(wall/112,2)
    labor_hours = round(wall/14,2)
    paint_cost = round(gallon * paint_price,2)
    labor_charge = round(labor_hours * 35,2)
    total = round(paint_cost + labor_charge)
    
    
    print(f'The number of gallons of paint required: {gallon}')
    print(f'The hours of labor required: {labor_hours}')
    print(f'The cost of the paint: {paint_cost}')
    print(f'The labor charges: {labor_charge}')
    print(f'The total cost of paint job: {total}')
    
paint_cal()
```

```python
def sales_tax():
    monthly_tax = float(input('Enter monthly sales tax:'))
    state = monthly_tax*0.05
    county = monthly_tax*0.025
    print(f'State sales tax: {state}\nCounty sales tax: {county}\
        \nTotal sales tax: {state + county}')
    
sales_tax()
```

```python
def feet_to_inches(a):
    #feet = float(input('Enter number of feet that you want to convert in inches: '))
    
    inches = a * 12
    return inches

feet = float(input('Enter number of feet that you want to convert in inches: '))
inch = feet_to_inches(feet)
print(f'Feet in inches: {inch}')
```

```python
def max(a,b):
    if a > b:
        temp = a
    else:
        temp = b
    return temp

first = int(input('Enter first integer value: '))
second = int(input('Enter second integer value: '))

greater = max(first,second)
print(f'greater value is: {greater}')
```

```python
def avg(a,b,c,d,e):
    return ((a+b+c+d+e)/5)
def grade(score):
    if 100 >= score >= 90:
        return 'A'
    if 89 >= score >= 80:
        return 'B'
    if 79 >= score >= 70:
        return 'C'
    if 69 >= score >= 60:
        return 'D'
    if score <= 50:
        return 'F'
    
first = int(input('Enter score of first test:'))
second = int(input('Enter score of second test:'))
third = int(input('Enter score of third test:'))
fouth = int(input('Enter score of fourth test:'))
fifth = int(input('Enter score of fifth test:')) 

average = avg(first,second,third,fouth,fifth)
print(f'average score of the tests: {average}')
print('*'*40)
test_score = int(input('Enter test score:'))

grd = grade(test_score)
print(f'Grade is: {grd}')
```
