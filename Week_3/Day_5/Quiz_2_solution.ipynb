```python
import numpy as np
from sklearn.datasets import load_digits
from numpy.linalg import norm

# Load the MNIST-like digits dataset
X = load_digits().data  # Each image is a vector of size (64,)

# Part 1: Cosine Similarity between the first and tenth image
u = X[0]  # First image
v = X[9]  # Tenth image

# Cosine similarity between u and v
cosine_similarity_uv = np.dot(u, v) / (norm(u) * norm(v))
print(f"Cosine similarity between first and tenth image: {cosine_similarity_uv}")

# Part 2: Efficient Cosine Similarity for All Image Combinations

# Compute the dot product matrix (pairwise dot products between all images)
dot_product_matrix = np.dot(X, X.T)

# Compute the magnitudes (Euclidean norms) for all images
magnitudes = norm(X, axis=1)

# Compute the outer product of magnitudes to form the denominator for cosine similarity
magnitude_matrix = np.outer(magnitudes, magnitudes)

# Element-wise division to compute cosine similarity matrix
cosine_similarity_matrix = dot_product_matrix / magnitude_matrix

print(f"Cosine similarity matrix shape: {cosine_similarity_matrix.shape}")
```

# Question 13

```python
import sympy as sp

# Define variables and constants
x = sp.symbols('x')
k, a, b = sp.symbols('k a b', constant=True)

# Function A: f(x) = x^3 - 1
f_A = x**3 - 1
derivative_A = sp.diff(f_A, x)

# Function B: f(x) = log(x^2 - 3k)
f_B = sp.log(x**2 - 3*k)
derivative_B = sp.diff(f_B, x)

# Function C: f(x) = exp(a * x^b)
f_C = sp.exp(a * x**b)
derivative_C = sp.diff(f_C, x)


print("Derivative of first function:", derivative_A)
print("Derivative of second function:", derivative_B)
print("Derivative of third function:", derivative_C)
```

# Question 14

```python
# Define variables
x, y = sp.symbols('x y')

# Function f(x, y) = exp(x^2 + 2*y^2)
f = sp.exp(x**2 + 2*y**2)

# First partial derivatives
f_x = sp.diff(f, x)  # fx
f_y = sp.diff(f, y)  # fy

# Second partial derivatives
f_xx = sp.diff(f_x, x)  # fxx
f_xy = sp.diff(f_x, y)  # fxy
f_yx = sp.diff(f_y, x)  # fyx
f_yy = sp.diff(f_y, y)  # fyy


print("Partial Derivative of f(x) = exp(𝑥 2 + 2 𝑦 2 ):", f_x)
print("Partial Derivative of f(y) = exp(𝑥 2 + 2 𝑦 2 ):", f_y)
print("Partial Derivative of f(xx) = exp(𝑥 2 + 2 𝑦 2 ):", f_xx)
print("Partial Derivative of f(yy) = exp(𝑥 2 + 2 𝑦 2 ):", f_yy)
print("Partial Derivative of f(xy) = exp(𝑥 2 + 2 𝑦 2 ):", f_xy)
print("Partial Derivative of f(yx) = exp(𝑥 2 + 2 𝑦 2 ):", f_yx)
```
